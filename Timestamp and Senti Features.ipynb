```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
header = [
    "Tweet Id", 
    "Username", 
    "Timestamp", 
    "#Followers",
    "#Friends",
    "#Retweets",
    "#Favorites",
    "Entities",
    "Sentiment",
    "Mentions",
    "Hashtags",
    "URLs"]

data = pd.read_csv("/Users/amarjyotkaur/Downloads/TweetsCOV19.tsv.gz", compression='gzip', names=header, sep='\t', quotechar='"')
```

```python
data[200:220]
```

```python
timecol = data['Timestamp']
timecol
```

```python
#Timestamp: Format ( "EEE MMM dd HH:mm:ss Z yyyy" ). ISOString => integer (e.g. 23517957).
import datetime
daykey = {"Sun":0,"Mon":1,"Tue":2,"Wed":3,"Thu":4,"Fri":5,"Sat":6}
monthkey = {"Jan":1, "Feb":2,"Mar":3,"Apr":4,"May":5,"Jun":6,"Jul":7,"Aug":8,"Sep":9,"Oct":10,"Nov":11,"Dec":12}
day = []
month = []
sec = [] #number of seconds from 1st Jan 2019
for i in timecol:
    t = i.split(" ")
    day.append(daykey[t[0]])
    month.append(monthkey[t[1]])
    date = datetime.datetime.strptime(i, '%a %b %d %H:%M:%S +0000 %Y')
    sec.append(str((date - datetime.datetime(2019, 1, 1)).total_seconds())[:-2])

#test
print(day[:10])
print(month[:10])
print(sec[:10])

```

```python
data['Day of Week'] = pd.Series(day)
data['Month'] = pd.Series(month)
data['Time Int'] = pd.Series(sec)
data
```

```python
# Tokenizing SentiStrength Score
data["Positive"] = pd.Series([ int(x.split(' ')[0]) for x in data['Sentiment']])
data["Negative"] = pd.Series([ int(x.split(' ')[1]) for x in data['Sentiment']])
data["Sentiment Disparity"] = pd.Series([int(x.split(' ')[0])-int(x.split(' ')[1]) for x in data['Sentiment']])
data
```

```python
#Entities 
entities = data['Entities'].str.split(";")
entity_no = []
for ent in entities:
    ent.pop()
    if ent[0]=='null':
        entity_no.append(0)
    else:
        entity_no.append(len(ent))
data['No. of Entities'] = entity_no
data.head()
```

```python
import logging
import numpy as np
import gensim.downloader

from gensim.models import Word2Vec
from gensim.test.utils import common_texts

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
#LOAD EMBEDDINGS
hashtag_embeddings = Word2Vec.load('/Users/amarjyotkaur/Downloads/hashtag_embeddings (1)')
mention_embeddings = Word2Vec.load('/Users/amarjyotkaur/Downloads/mention_embeddings')
hashtags = hashtag_embeddings.wv.index_to_key
mentions = mention_embeddings.wv.index_to_key

mentions
```

```python
mention_embeddings.wv['realDonaldTrump']
```

```python
ls = []
for i in data['Mentions']:
    if i in mentions and i!="null;":
        ls.append(mention_embeddings.wv[i])
    else:
        ls.append(np.zeros((25,), dtype=int))
data['Mention Embedding'] = pd.Series(ls)
```

```python
ls = []
for i in data['Hashtags']:
    if i in hashtags and i!="null;":
        ls.append(hashtag_embeddings.wv[i])
    else:
        ls.append(np.zeros((25,), dtype=int))
data['Hashtag Embedding'] = pd.Series(ls)
```
